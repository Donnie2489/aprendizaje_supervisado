```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x='Survived')
plt.show()
```

La distribución de la variable objetivo (Survived) muestra que la mayoría de los pasajeros no sobrevivió al naufragio. Aproximadamente el 61.6% de los pasajeros falleció, mientras que el 38.4% sobrevivió. Esto indica un ligero desbalance entre las clases.

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.show()
```
